```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP



# Assuming 'tcrdist' is your computed distance matrix
tcrdist_matrix = dmats['tcrdist']

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tcrdist_matrix)
df_beta1['pca1'] = pca_result[:, 0]
df_beta1['pca2'] = pca_result[:, 1]

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(tcrdist_matrix)
df_beta1['tsne1'] = tsne_result[:, 0]
df_beta1['tsne2'] = tsne_result[:, 1]

# UMAP
umap = UMAP(n_components=2)
umap_result = umap.fit_transform(tcrdist_matrix)
df_beta1['umap1'] = umap_result[:, 0]
df_beta1['umap2'] = umap_result[:, 1]
```

```python
# Plotting
plt.figure(figsize=(15, 5))

# PCA plot
plt.subplot(1, 3, 1)
sns.scatterplot(x='pca1', y='pca2', hue='color', data=df_beta1)
plt.title('PCA')
# Move the legend outside the plot area
plt.legend(title='Antigen Species', bbox_to_anchor=(0.5, -0.2))

# t-SNE plot
plt.subplot(1, 3, 2)
sns.scatterplot(x='tsne1', y='tsne2', hue='color', data=df_beta1)
plt.title('t-SNE')
# Move the legend outside the plot area
plt.legend(title='Antigen Species', bbox_to_anchor=(0.5, -0.2))

# UMAP plot
plt.subplot(1, 3, 3)
sns.scatterplot(x='umap1', y='umap2', hue='color', data=df_beta1)
plt.title('UMAP')

# Move the legend outside the plot area
plt.legend(title='Antigen Species', bbox_to_anchor=(0.5, -0.2))

plt.show()
```
